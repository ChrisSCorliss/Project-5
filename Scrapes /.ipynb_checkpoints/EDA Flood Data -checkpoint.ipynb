```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, statsmodels.api as sm
from sklearn.model_selection import train_test_split
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import adfuller

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
con  = pd.read_csv('../Datasets/concrete.csv')
mv = pd.read_csv('../Datasets/mount_vernon.csv')
```

```python
con['timezone'] = con['timezone'].replace('PDT', 'PST')
mv['timezone'] = mv['timezone'].replace('PDT', 'PST')
```

```python
con['datetime'] = pd.to_datetime(con['datetime'])
```

```python
mv['datetime'] = pd.to_datetime(mv['datetime'])
```

```python
print(mv.dtypes)

mv.head()
```

```python
mv['discharge'] = mv['discharge'].astype('int64')

print(mv.dtypes)

mv.head(1)
```

```python
print(mv.head())
print(mv.tail())
print(con.head())
print(con.tail())
```

```python
combined = mv.merge(con, how = 'outer', on = 'datetime')
```

```python
np.sum(combined.isna())
```

```python
combined = combined.dropna()
```

```python
combined.head()
```

```python
con = combined[['city_y', 'datetime', 'timezone_y', 'discharge_y', 'gauge_lvl_y']]
mv = combined[['city_x', 'datetime', 'timezone_x', 'discharge_x', 'gauge_lvl_x']]
```

```python
con.head()
```

```python
mv.head()
```

```python
con = con.rename(columns = {'city_y': 'city',
                      'timezone_y': 'timezone', 
                      'discharge_y': 'discharge',
                      'gauge_lvl_y': 'gauge_lvl'})
```

```python
mv = mv.rename(columns = {'city_x': 'city',
                     'timezone_x': 'timezone', 
                     'discharge_x': 'discharge', 
                     'gauge_lvl_x': 'gauge_lvl'})
```

```python
print(con.head())
print(con.tail())
```

```python
print(mv.head())
print(mv.tail())
```

```python
con.to_csv('../Datasets/cleaned_concrete_water.csv',index = False)
mv.to_csv('../Datasets/cleaned_mount_vernon_water.csv', index = False)
```

```python
fig, axis = plt.subplots(nrows = 3, figsize = (15,12))
axis[0].plot(con['gauge_lvl'])
axis[1].plot(con['gauge_lvl'].diff())
axis[2].plot(con['gauge_lvl'].pct_change())
axis[0].set_title('Concrete River Gauge Level')
axis[1].set_title('Concrete River Gauge Level Differenced')
axis[2].set_title('Concrete River Gauge Level Percent Change')
plt.savefig('Concrete');
```

```python
fig, axis = plt.subplots(nrows = 3, figsize = (15,12))
axis[0].plot(mv['gauge_lvl'])
axis[1].plot(mv['gauge_lvl'].diff())
axis[2].plot(mv['gauge_lvl'].pct_change())
axis[0].set_title('Mount Vernon River Gauge Level')
axis[1].set_title('Mount Vernon River Gauge Level Differenced')
axis[2].set_title('Mount Vernon River Gauge Level Paercent Change')
plt.savefig('Mount_Vernon');
```
